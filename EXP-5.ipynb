```python
import numpy as np
array=np.random.randint(1,150,20)
def outDetection(array):
    sorted(array)   
    Q1,Q3=np.percentile(array,[25,75])
    IQR=Q3-Q1
    lr=Q1-(1.5*IQR)
    ur=Q3+(1.5*IQR)
    return lr,ur
lr,ur=outDetection(array)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.displot(array)
plt.show()
```

```python
sns.distplot(array)
plt.show()
```

```python
new_array=array[(array>lr) & (array<ur)]
sns.displot(new_array);
plt.show()
```

```python
sns.distplot(new_array);
plt.show()
```
